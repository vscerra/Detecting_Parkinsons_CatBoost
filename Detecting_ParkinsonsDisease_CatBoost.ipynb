## Importing the libraries

```python
! pip install catboost
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, confusion_matrix
from catboost import CatBoostClassifier
```

## Importing the dataset


```python
dataset = pd.read_csv('/content/parkinsons.data')
dataset.head()
```

```python
y = dataset.loc[:,'status'].values
X = dataset.loc[:,dataset.columns != 'status'].values[:,1:]
#Get the counts of each label (0 and 1)
print(y[y==1].shape[0], y[y==0].shape[0])
```

## Scale the data

```python
#Scale features to be between -1 and 1
scaler = MinMaxScaler((-1, 1))
X = scaler.fit_transform(X)
```

## Splitting the dataset into training and test sets

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .20)
```

## Training XGBoost on the training set

```python
#classifier = XGBClassifier()
classifier = CatBoostClassifier()
classifier.fit(X_train, y_train)
```

## Making a confusion matrix

```python
y_pred = classifier.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

## Applying K-fold cross validation

```python
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 10)
print('Accuracy: {:.2f} %'.format(accuracies.mean()*100))
print('Standard deviation: {:.2f} %'.format(accuracies.std()*100))
```
